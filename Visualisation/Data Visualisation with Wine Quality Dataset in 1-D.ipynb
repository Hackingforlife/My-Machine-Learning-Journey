```python
# Read the data
white_wine=pd.read_csv('winequality-white.csv', sep=';')
red_wine=pd.read_csv('winequality-red.csv', sep=';')
```

```python
# Store wine type as an attribute
red_wine['wine_type']='red'
white_wine['wine-type']='white'
```

```python
#Assigning the quality of wine into qualities
red_wine.loc[red_wine.quality<5, 'quality_label']='low'
```

```python
red_wine.loc[(red_wine.quality>=5), 'quality_label']='medium'
red_wine.loc[(red_wine.quality>=7), 'quality_label']='high'
```

```python
red_wine.head()
```

```python
white_wine.loc[(white_wine.quality<5), 'quality_label']='low'
white_wine.loc[(white_wine.quality>=5), 'quality_label']='medium'
white_wine.loc[(white_wine.quality>=7), 'quality_label']='high'
```

```python
white_wine.head(20)
```

```python
#merge the white_wine and red_wine datasets
wine=pd.concat([red_wine, white_wine])
#shuffling so as to have random data points
wine=wine.sample(frac=1, random_state= 42).reset_index(drop=True)
```

```python
subset_attributes=['residual sugar', 'total sulfur dioxide', 'sulphates', 'alcohol', 'volatile acidity', 'quality']
rs=round (red_wine[subset_attributes].describe(),2)
ws=round (white_wine[subset_attributes].describe(),2)

pd.concat([rs,ws], axis=1, keys=['Red Wine Statistics', 'White Wine Statistics'])
```

```python
#The Tableau Color Scheme
tableau20=[(31,119,180),(174,199,232),(255,127,14), (255,187,120),
          (44,160,44), (152,223,138), (214,39,40), (255,152,150),
          (148,103,189), (197,176,213), (140,86,75), (196,156,148),
          (227,119,194), (247,182, 210),(127,127,127), (199,199,199),
          (188,189,34), (219,219, 141), (23,190,207), (158,218,229)]

for eachset_of_colors in range(len(tableau20)):
    r,g,b= tableau20[eachset_of_colors]
    tableau20[eachset_of_colors]=(r/255., g/255., b/255.)
    
#1-Dimension Visualisation
wine.hist(bins=15, color=tableau20[10], edgecolor=tableau20[15], linewidth=10, xlabelsize=8, ylabelsize=8, grid=False)
plt.tight_layout(rect=(0,0,1.2, 1.2))
```

```python
#Histogram Plot
fig=plt.figure(figsize=(6,4))
title=fig.suptitle("Sulphates Content in Wine", fontsize=14)
fig.subplots_adjust(top=0.85, wspace=0.3)

ax=fig.add_subplot(1,1, 1)
ax.set_xlabel("Sulphates")
ax.set_ylabel("Frequency")
ax.text(1.2, 800, r'$\mu$ ='+str(round(wine['sulphates'].mean(),2)),fontsize=12)

freq,bins,patches=ax.hist(wine['sulphates'], color=tableau20[14], bins=15, edgecolor=tableau20[9], linewidth=1)
```

```python
# Density plot
fig=plt.figure(figsize=(6,4))
title=fig.suptitle("Sulphates Content in wine", fontsize=14)
fig.subplots_adjust(top=0.85, wspace=0.3)

ax1=fig.add_subplot(1,1, 1)
ax1.set_xlabel('Sulphates')
ax1.set_ylabel('Frequency')
sns.kdeplot(wine['sulphates'], ax=ax1, shade=True, color=tableau20[9])
```

```python
#Analysing one Attribute
fig=plt.figure(figsize=(6,4))
title=fig.suptitle("Wine Quality Frequncy", fontsize=14)
fig.subplots_adjust(top=0.85, wspace=0.3)

ax=fig.add_subplot(1,1, 1)
ax.set_xlabel("Quality")
ax.set_ylabel("Frequency")

w_q=wine['quality'].value_counts()
w_q=(list(w_q.index), list(w_q.values))

ax.tick_params(axis='both', which='major', labelsize=8.5)
bar=ax.bar(w_q[0], w_q[1], color=tableau20[8])
```
